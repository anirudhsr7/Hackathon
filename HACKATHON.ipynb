```python
#ANIRUDH AND AKARSHIT


class SocialNetworking:
    
    def __init__(self, year, users):
        self.year = year
        self.users = users
    
    def growth(self):
        self.growths = [] 
        for i in range(len(self.users)-1):
            ratio = (self.users[i+1] - self.users[i]) / (self.users[i+1] + self.users[i])
            ratio *= 100
            self.growths.append(ratio)
        return self.growths
            
        
```

```python
from matplotlib import pyplot as plt
```

```python
facebook_years = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
facebook_users = [197, 360, 608, 845, 1056, 1228, 1393, 1591, 1860, 2129, 2196]

snapchat_years = [2014, 2015, 2016, 2017, 2018]
snapchat_users = [71, 107, 158, 187, 191]

instagram_years = [2013, 2014,2015, 2016, 2017, 2018]
instagram_users = [150, 300, 400, 600, 800, 1000]

facebook = SocialNetworking(facebook_years,facebook_users)
f = facebook.growth()
facebook_years_c = ['2008-2009', '2009-2010', '2010-2011', '2011-2012', '2012-2013', '2013-2014', '2014-2015', '2015-2016', '2016-2017', '2017-2018']

snap = SocialNetworking(snapchat_years, snapchat_users)
s = snap.growth()
snap_years_c = ['2014-2015', '2015-2016', '2016-2017', '2017-2018']

insta = SocialNetworking(instagram_years, instagram_users)
i = insta.growth()
insta_years_c = ['2013-2014','2014-2015', '2015-2016', '2016-2017', '2017-2018']
```

```python
plt.bar(facebook_years, facebook_users,label="Facebook",width=0.5)
plt.ylabel("Users")
plt.legend()
plt.show()

```

```python
plt.bar(snapchat_years, snapchat_users,label="Snapchat",width=0.5)
plt.legend()
plt.ylabel("Users")
plt.show()
```

```python
plt.bar(instagram_years, instagram_users,label="Instagram",width=0.5)
plt.legend()
plt.ylabel("Users")
plt.show()
```

```python

plt.plot(facebook_years_c, f,'g',label='facebook',linewidth=3)
plt.plot(snap_years_c, s,'r',label='snapchat',linewidth=3)
plt.plot(insta_years_c, i,'hotpink',label='instagram',linewidth=3)
plt.ylabel("Users")
fig=[60,10]
plt.rcParams["figure.figsize"]= fig
plt.legend()
plt.show()

```
